```python
# ========================
# 1. Setup & Imports
# ========================
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

sns.set(style="whitegrid")
```

```python
# ========================
# 2. Upload Data
# ========================
print("Upload matches.csv")
uploaded = files.upload()

print("Upload player_stats.csv")
uploaded = files.upload()

# Load datasets
matches = pd.read_csv("matches.csv")
player_stats = pd.read_csv("player_stats.csv")
```

```python
# ========================
# 3. Explore Data (EDA)
# ========================
print("Matches dataset:")
display(matches.head())
matches.info()

print("\nPlayer Stats dataset:")
display(player_stats.head())
player_stats.info()

# Quick stats
print("\nMatch summary:")
display(matches.describe())

print("\nPlayer stats summary:")
display(player_stats.describe())
```

```python
# ========================
# 4. Team-Level Analysis
# ========================

# Average points scored at home
avg_home_scores = matches.groupby("HomeTeam")["HomeScore"].mean().sort_values(ascending=False)
print("Average home scores:")
print(avg_home_scores)

# Average points scored away
avg_away_scores = matches.groupby("AwayTeam")["AwayScore"].mean().sort_values(ascending=False)
print("\nAverage away scores:")
print(avg_away_scores)

# Visualization: Average home scores
plt.figure(figsize=(10,6))
sns.barplot(x=avg_home_scores.index, y=avg_home_scores.values, palette="Blues_d")
plt.xticks(rotation=45)
plt.title("Average Points Scored at Home by Team")
plt.ylabel("Average Score")
plt.xlabel("Team")
plt.show()

# Visualization: Average away scores
plt.figure(figsize=(10,6))
sns.barplot(x=avg_away_scores.index, y=avg_away_scores.values, palette="Blues_d")
plt.xticks(rotation=45)
plt.title("Average Points Scored Away by Team")
plt.ylabel("Average Score")
plt.xlabel("Team")
plt.show()
```

```python
# ========================
# 5. Player-Level Analysis
# ========================

# Top 10 try scorers
top_tries = player_stats.groupby("PlayerName")["Tries"].sum().sort_values(ascending=False).head(10)
print("Top 10 Try Scorers:")
print(top_tries)

plt.figure(figsize=(10,6))
sns.barplot(x=top_tries.values, y=top_tries.index, palette="Greens_d")
plt.title("Top 10 Try Scorers")
plt.xlabel("Total Tries")
plt.ylabel("Player")
plt.show()

# Top 10 tacklers
top_tackles = player_stats.groupby("PlayerName")["Tackles"].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Tacklers:")
print(top_tackles)

plt.figure(figsize=(10,6))
sns.barplot(x=top_tackles.values, y=top_tackles.index, palette="Reds_d")
plt.title("Top 10 Tacklers")
plt.xlabel("Total Tackles")
plt.ylabel("Player")
plt.show()
```

```python
# ========================
# 6. Save Clean Outputs
# ========================

# Player summary (aggregated stats)
player_summary = player_stats.groupby("PlayerName").sum()
player_summary.to_csv("player_summary.csv", index=True)

# Download to local machine
files.download("player_summary.csv")
```

```python
import numpy as np

# If attendance & ticket price don't exist, simulate them
if "attendance" not in matches.columns:
    np.random.seed(42)
    matches["attendance"] = np.random.randint(5000, 25000, size=len(matches))  # between 5k25k spectators

if "ticket_price" not in matches.columns:
    matches["ticket_price"] = np.random.randint(50, 250, size=len(matches))  # between R50R250 per ticket
```

```python
matches["revenue"] = matches["attendance"] * matches["ticket_price"]
matches[["HomeTeam", "attendance", "ticket_price", "revenue"]].head()
```

```python
team_revenue = matches.groupby("HomeTeam")["revenue"].sum().sort_values(ascending=False)
print("Total Revenue by Team:")
print(team_revenue)

# Visualization
import matplotlib.pyplot as plt
sns.set(style="whitegrid") # Ensure style is set for visualization
plt.figure(figsize=(10,6))
sns.barplot(x=team_revenue.index, y=team_revenue.values, palette="Purples_d")
plt.xticks(rotation=45)
plt.title("Total Ticket Revenue by Team (Home Games)")
plt.ylabel("Revenue (ZAR)")
plt.xlabel("Team")
plt.show()
```

```python
# Determine match winner
def get_winner(row):
    if row["HomeScore"] > row["AwayScore"]:
        return row["HomeTeam"]
    elif row["AwayScore"] > row["HomeScore"]:
        return row["AwayTeam"]
    else:
        return "Draw"

matches["winner"] = matches.apply(get_winner, axis=1)
matches[["HomeTeam", "AwayTeam", "HomeScore", "AwayScore", "winner"]].head()
```

```python
wins = matches[matches["winner"] != "Draw"].groupby("winner").size()
wins = wins.rename("wins").reset_index()
wins.head()
```

```python
team_revenue_df = team_revenue.reset_index().rename(columns={"revenue": "total_revenue"})
team_performance = pd.merge(wins, team_revenue_df, left_on="winner", right_on="HomeTeam", how="left")

team_performance = team_performance[["winner", "wins", "total_revenue"]]
display(team_performance.head())
```

```python
correlation = team_performance["wins"].corr(team_performance["total_revenue"])
print(f"Correlation between wins and total revenue: {correlation:.2f}")
```

```python
plt.figure(figsize=(8,6))
sns.regplot(
    data=team_performance,
    x="wins",
    y="total_revenue",
    scatter_kws={"s": 80, "color": "blue"},
    line_kws={"color": "red"}
)
plt.title("Relationship Between Wins and Ticket Revenue")
plt.xlabel("Wins")
plt.ylabel("Total Revenue (ZAR)")
plt.show()
```
